# Naive Bayes for Data Discrete: Dữ liệu rời rạc

```python
import numpy as np
# Navie Bayes classification
class NB():
  def fit(self,X,y):
    self.X = X
    self.Label = y
    # Lấy ra các giá trị đơn trị của nhãn
    self.Label_unique = np.unique(y)
    # Tính xác suất cho từng nhãn
    self.P_Label = np.array([(self.Label[self.Label==value].size/self.Label.size ) for value in self.Label_unique])
  def predict(self,x_pred):
    P_result = np.array([np.prod(((self.X[self.Label == value])== x_pred).sum(axis=0)/(self.Label[self.Label == value]).size)  for value in self.Label_unique])*self.P_Label
    print(P_result)
    return self.Label_unique[np.argmax(P_result)] # Kết quả phân lớp lấy ra nhãn có xác suất là lớn nhất
```

# Gaussian Naive Bayes for classification with Continuous Data: Dữ liệu liên tục

```python
class GaussianNB():
  def fit(self,X,y):
    self.X= X
    self.Label = y
    self.Label_unique = np.unique(self.Label)
    self.P_Label = np.array([self.Label[self.Label==value].size/self.Label.size for value in self.Label_unique])
  def predict(self,x_pred):
    # Tính mean cho từng feature khi Label = ... Xảy ra
    self.mean = np.array([self.X[self.Label==value].mean(axis=0) for value in self.Label_unique])
    # Tính variance(Phương sai bình phương) cho từng feature khi Label= ... xảy ra
    self.var = np.array([self.X[self.Label==value].var(axis=0) for value in self.Label_unique])

    self.P_result = np.prod((1/np.sqrt(2*np.pi*self.var))*np.exp((-(x_pred-self.mean)**2)/(2*self.var)),axis=1).flatten()*self.P_Label
    return self.Label_unique[np.argmax(self.P_result)]
```

```python
FILE_PATH3 = '/content/drive/MyDrive/AIO 2024/Module_2/Week-3/CN/iris.data.csv'
```

```python
df = df.to_numpy()
df
```

```python
Label = df[:,-1].astype(str)
print(f"Label: {Label}")
X_train = df[:,:-1].astype(float)
print(f"X_train: {X_train}")
```

```python
# Mã hóa Nhãn
Label_unique = np.unique(Label)
print(f"Label_unique: {Label_unique}")
for i in range(Label_unique.size):
  Label = np.char.replace(Label,Label_unique[i],str(i))
Label= Label.astype(float)
```

```python
print(Label)
```

```python
# Khởi tạo mô hình
clf = GaussianNB()
```

```python
# Huấn luyện mô hình
clf.fit(X_train,Label)
```

```python
# Dự đoán mô hình
clf.predict(np.array([4.9 ,3.1 ,1.5 ,0.1]))
```
